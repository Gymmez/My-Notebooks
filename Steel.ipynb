```python
df=pd.read_csv('/kaggle/input/steel-dataset/Steel_industry.csv')
X=df.drop(['Date_Time','WeekStatus','Day_Of_Week'],axis=1)
target=X.pop('Load_Type')
target=target.replace('Light_Load',0)
target=target.replace('Medium_Load',0)
target=target.replace('Maximum_Load',0)
target
```

```python
mutual=mutual_info_classif(X, target)
mutual_info=pd.Series(mutual)
mutual_info.index=X.columns
mutual_info.sort_values(ascending=False)

target=np.array(target)
X=np.array(X)
```

```python
model=tf.keras.Sequential()
model.add(tf.keras.layers.Dense(128, activation='relu'))
model.add(tf.keras.layers.Dense(3, activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
model.fit(X,target, epochs=2, validation_split=0.2)
```

```python
model.save('/kaggle/working/steel_load.h5')
```
